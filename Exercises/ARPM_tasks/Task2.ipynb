```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
```

```python
# Define the function
def f(x):
    return 0.25 * (x + 4) * (x + 1) * (x - 2)

# Define the x range
x = np.linspace(-4, 4)

# Set the figure size
plt.figure(figsize=(1280/72, 720/72))

# Plot the function
plt.plot(x, f(x), color='#539692')

# Find the minimum of the function
x_min = x[np.argmin(f(x))]
y_min = f(x_min)

# Annotate the minimum
plt.annotate(f'Minimum', xy=(x_min, y_min),
             xytext=(-30, 40), textcoords='offset points', fontsize=12,
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Add the embedded figure
ax = plt.axes([0.6, 0.6, 0.25, 0.25])
ax.plot(x, f(x), color='#539692')
ax.set_xlim(-2.7, 2)
ax.set_ylim(-2.7, 2)

# Show the plot
plt.show()

# Save the plot
plt.savefig('Task2Function.png')
```

```python
# Set the figure size
fig = plt.figure(figsize=(1280/72, 720/72))

# Define the function
def f(x):
    return 0.25 * (x + 4) * (x + 1) * (x - 2)

# Define the x range
x = np.linspace(-4, 4, 100)

# Plot the function
plt.plot(x, f(x), color='#539692')

# Find the minimum of the function
x_min = x[np.argmin(f(x))]
y_min = f(x_min)

# Annotate the minimum
plt.annotate(f'Minimum', xy=(x_min, y_min),
             xytext=(-30, 40), textcoords='offset points', fontsize=12,
             arrowprops=dict(facecolor='black', arrowstyle='->'))

# Animation
plt.rcParams['animation.ffmpeg_path']='C:/FFmpeg/bin/ffmpeg.exe'

# Add a dot at the starting point
dot, = plt.plot(x[0], f(x[0]), 'ro', markersize=10)

# Add the embedded figure
ax = plt.axes([0.6, 0.6, 0.25, 0.25])
ax.plot(x, f(x), color='#539692')
ax.set_xlim(-2.7, 2)
ax.set_ylim(-2.7, 2)

# Define the update function for the animation
def update(frame):
    # Move the dot to the next position
    dot.set_data(x[frame], f(x[frame]))
    return dot,

# Create the animation
ani = FuncAnimation(fig, update, frames=len(x), interval=50, blit=True)

# Save the animation in mp4 format
ani.save("Animation.mp4", dpi=72)
```
