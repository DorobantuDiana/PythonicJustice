```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
m = [11, 11]
sigma = [[1, 0.2], [0.2, 0.8]]
sample_size = 10000

sample = np.random.multivariate_normal(m, sigma, size=sample_size)
print(sample)
```

```python
x = np.array(sample[:,0])
y = np.array(sample[:,1])
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(x, y, color='#1e406c')
plt.xlabel("Variable X1")
plt.ylabel("Variable X2")
plt.title("Distribution", size=18)
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
plt.show()

```

```python
plt.figure(figsize=(10, 10))

# create a grid of 3 rows and 3 columns
gs = gridspec.GridSpec(3, 3)

# create a subplot at the first and second rows, and the second and third columns
ax_main = plt.subplot(gs[0:2, 1:3])

# create a subplot at the third row, and the first and second columns, sharing the x-axis with ax_main
ax_xDist = plt.subplot(gs[2, 1:3],sharex=ax_main)

# # create a subplot at the first and second rows, and the first column, sharing the y-axis with ax_main
ax_yDist = plt.subplot(gs[0:2, 0],sharey=ax_main)

ax_main.scatter(x,y,marker='.', color='#1e406c')
ax_main.set(xlabel="Variable X1", ylabel="Variable X2")

ax_xDist.hist(x,bins=100,align='mid', color='#1e406c')
ax_xDist.set(ylabel='')

ax_yDist.hist(y,bins=100,orientation='horizontal',align='mid', color='#1e406c')
ax_yDist.set(xlabel='')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation

# Set the figure size
fig = plt.figure(figsize=(1280/72, 720/72))

# Define the function
def f(x):
    return 0.25 * (x + 4) * (x + 1) * (x - 2)

# Define the x range
x = np.linspace(-4, 4, 100)

# Plot the function
plt.plot(x, f(x), color='#539692')

# Find the minimum of the function
x_min = x[np.argmin(f(x))]
y_min = f(x_min)

# Annotate the minimum
plt.annotate(f'Minimum', xy=(x_min, y_min),
             xytext=(-30, 40), textcoords='offset points', fontsize=12,
             arrowprops=dict(facecolor='black', arrowstyle='->'))


plt.rcParams['animation.ffmpeg_path']='C:/FFmpeg/bin/ffmpeg.exe'
# Add a dot at the starting point
dot, = plt.plot(x[0], f(x[0]), 'ro', markersize=10)

# Add the embedded figure
ax = plt.axes([0.6, 0.6, 0.25, 0.25])
ax.plot(x, f(x), color='#539692')
ax.set_xlim(-2.7, 2)
ax.set_ylim(-2.7, 2)

# Define the update function for the animation
def update(frame):
    # Move the dot to the next position
    dot.set_data(x[frame], f(x[frame]))
    return dot,

# Create the animation
ani = FuncAnimation(fig, update, frames=len(x), interval=50, blit=True)
# Save the animation in mp4 format
ani.save("Animation.mp4", dpi=72)



```
